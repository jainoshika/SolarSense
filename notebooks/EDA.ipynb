```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    "../data/processed/solar_features.csv"
)

df.head()
```

```python
print("Shape:",df.shape)

df.info()

df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12,7))

sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title(
    "Feature Correlation"
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    x="Hour",
    y="Power",
    data=df
)

plt.title(
    "Power vs Hour"
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(
    x="AMBIENT",
    y="Power",
    data=df
)

plt.title(
    "Ambient Temperature vs Power"
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x="SEASON",
    y="Power",
    data=df
)

plt.title(
    "Season-wise Power Distribution"
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["Power"],
    kde=True
)

plt.title(
    "Power Distribution"
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x="TimePeriod",
    y="Power",
    data=df
)

plt.title(
    "Power by Time Period"
)

plt.show()
```
